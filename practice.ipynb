# 3. Categorical plots

## 3.1. Bar plot

```python
tips_df.head()
```

```python
sns.set_style('darkgrid')

sns.set_context(context='paper', font_scale=1.4)

sns.barplot(x='sex', y='total_bill', data=tips_df)  # estimates by mean by default
```

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.median)
```

## 3.2. Count plot

```python
sns.countplot(x='sex', data=tips_df)
```

## 3.3. Box plot

```python
sns.boxenplot(x='day', y='total_bill', data=tips_df, hue='sex')

plt.gcf().set_size_inches(13.5, 7)
```

## 3.4. Violin plot

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split=True)
```

## 3.5. Strip plot

```python
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', jitter=True, dodge=True)
plt.gcf().set_size_inches(13.5, 7)
plt.legend(loc='lower right')
```

## 3.6. Swarm plot

```python
sns.swarmplot(x='day', y='total_bill', data=tips_df)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips_df)
sns.swarmplot(x='day', y='total_bill', data=tips_df, color='black')
```

# 4. Matrix plots

## 4.1. Heatmaps

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights = flights.pivot_table(values='passengers', columns='year', index='month')
flights.head()
```

```python
sns.heatmap(flights, cmap='Blues', linecolor='white', linewidths=1)
plt.xticks(rotation=45)
```
